```python
import os
from os import path
from wordcloud import WordCloud


d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()
text = open(path.join(d, '/Users/soumilhooda/Desktop/Dad_PapersForWordCloud/combinedDadPapers.txt')).read()

wordcloud = WordCloud().generate(text)

import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig('dad1.png')


wordcloud = WordCloud(max_font_size=100, max_words=100, background_color="white", relative_scaling=1, min_word_length=5 , scale=2, normalize_plurals=True).generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation='gaussian')
plt.axis("off")
plt.savefig('dad2.png')
plt.show()
```
